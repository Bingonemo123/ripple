```python
from iqoptionapi.stable_api import IQ_Option
connector =IQ_Option("ww.bingonemo@gmail.com","JF*#3C5va&_NDqy")
```

```python
TrendLib = []
with open(r"C:\Users\MSI\Downloads\c.log", "r+", encoding="utf-8") as file:
    for x in file.readlines():
        for ch in range(len(x)):
            if x[ch] == " ":
                TrendLib.append(x[:ch -2])
                break

def CONNECTIONCHECK ():
    if connector.check_connect()==False:#detect the websocket is close
        print("try reconnect")
        check,reason=connector.connect()         
        if check:
            print("Reconnect successfully")
    return connector.check_connect()
        

def DECISIONMAKER (PastData):
    TFPastData = PastData
    for trend in TrendLib:
        if len(trend) > len(TFPastData):
            continue
        for (num, ChrTF) in zip(range(len(trend)), reversed(trend)):
            if ChrTF == "*":
                continue
            if ChrTF != "↓↑"[TFPastData[-1-num]]: 
                break
        else:
            return False, trend
            
    return None, 'No Trend'
```

```python
goal="EURUSD"
size=30
maxdict=20
def dataREQUEST():
    
    candles=connector.get_realtime_candles(goal,size)

    TFfx = [(candles[timstm]['close'] - candles[timstm]["open"] ) > 0 for timstm in candles]
    return TFfx

def BROKER ():
    ACTIVES = "EURUSD"
    ACTION = "put"
    MONEY = 1
    expiration_mode =0.5
    check,id=connector.buy(MONEY,
                ACTIVES,
                ACTION,
                expiration_mode)

```

```python
import time

def TimeGate ():
    while time.time()%60 > 3:
        pass
    return True
#Connection Setup While Loop
while True:
    connector.connect()
    connector.change_balance("PRACTICE")

    connector.start_candles_stream("EURUSD", size, maxdict)
    # Working Loop
    while TimeGate() and CONNECTIONCHECK() :
        print(f'Running Tests {time.strftime("%H:%M:%S", time.localtime())}')
        TFfx = dataREQUEST()
        PHROPET = DECISIONMAKER(TFfx)
        print(PHROPET)

        time.sleep(28)

        if PHROPET[0] == False:
            CONNECTIONCHECK()
            BROKER()
    connector.stop_candles_stream("EURUSD", size)
```

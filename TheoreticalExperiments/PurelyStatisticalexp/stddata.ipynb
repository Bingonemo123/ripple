```python
import meanvis
import numpy as np
import pandas as pd
```

```python
%matplotlib widget
meanvis.rates_frame['close'].plot()
```

```python
df = pd.DataFrame([meanvis.rates_frame['close'][:i].mean(axis=0) for i in range(len(meanvis.rates_frame['close']))])
df.plot()
```

```python
dfs = pd.DataFrame([df[1:i].mean(axis=0) for i in range(2, len(df))])
dfs.plot()
```

```python
stdf = pd.DataFrame([meanvis.rates_frame['close'][:i].std(axis=0) for i in range(len(meanvis.rates_frame['close']))])
stdf.plot()
```

```python
stdfs = pd.DataFrame([stdf[1:i].mean(axis=0) for i in range(2, len(stdf))])
stdfs.plot()
```

```python
stdfs = pd.DataFrame([stdfs[1:i].mean(axis=0) for i in range(2, len(stdfs))])
stdfs.plot()
```

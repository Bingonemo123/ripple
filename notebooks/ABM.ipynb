```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import inflect

# Create an engine for converting numbers to words
InfEngine = inflect.engine()

residual_df = pd.read_csv(
    "ROKU_rand60_residuals.csv",
    index_col=0,
    parse_dates=True,
    date_format="%H:%M:%S"
)
residual_df.index = residual_df.index.time

seasonal_df = pd.read_csv(
    "ROKU_rand60_seasonal.csv",
    index_col=0,
    parse_dates=True,
    date_format="%H:%M:%S"
)
seasonal_df.index = seasonal_df.index.time

### Normalize the seasonal data ###
''' Make it greater than zero and 
add small value to get rid of zero'''
if seasonal_df.min().item() < 0 :
    seasonal_df += abs(seasonal_df.min())

min_idx = seasonal_df['seasonal'].idxmin()

if seasonal_df.loc[min_idx, 'seasonal'] == 0:
    seasonal_df.loc[min_idx, 'seasonal'] += 1e-6

beta_df = (residual_df['resid']*(number_of_agents + 1))\
    /(2*seasonal_df['seasonal']) # calculating betas for betabinomial
```

```python
from scipy.stats import betabinom


for t in seasonal_df.index:
    agents_trading = betabinom.rvs(n=number_of_agents, a=seasonal_df.loc[t], b=beta_df.loc[t])
    print(t, agents_trading)
```
